## Section 1 — Making a Copy

When you open someone else's notebook, it's **read-only**. Making a copy saves it to your own Google Drive so you can edit and save changes.

**How to make a copy:**
1. Click **File → Save a copy in Drive** from the top menu.
2. A new tab will open with *your own editable copy* — it's automatically named **"Copy of [original name]"**.
3. You can rename it by clicking on the title at the top of the page.

> **Always** make a copy before editing a shared notebook!

## Section 2 — Saving Your Work

Colab **auto-saves** your notebook to Google Drive every few minutes, but you can also save manually.

**Saving options:**
- **Auto-save:** Happens automatically in the background.
- **Manual save:** Press `Ctrl + S` (Windows/Linux) or `Cmd + S` (Mac), or go to **File → Save**.
- **Save a copy in Drive:** File → Save a copy in Drive (creates a new file).
- **Download:** File → Download → Download .ipynb (saves a local copy to your computer).

> All saved notebooks appear in your Google Drive under the **Colab Notebooks** folder. You can move the Colab file to a different location within Drive and not see a change in function.

## Section 3 — Text Blocks & Code Blocks

Colab notebooks are made up of two types of **cells** (blocks):

### Text (Markdown) Cells
Used for writing explanations, headings, and notes. They support **Markdown** formatting:
- `# Heading 1`, `## Heading 2` — for headings

Double-click any text cell to **edit** it. Click outside or press `Shift + Enter` to render it.

**Adding new cells:**
- Hover between cells to see **+ Code** and **+ Text** buttons.
- Or use the top toolbar buttons which say **+ Code** and **+ Text**.

## Section 4 — Runtime Type

A **runtime** is the computing environment that actually runs your code. Colab gives you a choice of hardware. The Free plan limits the connection time and types of hardware available, paid plans increase your options.

> For most activities in BMB, the free Colab plan is sufficient.

**How to set your runtime type:**
1. In the top menu, click **Runtime → Change runtime type**.
2. A dialog box will appear with a **Hardware accelerator** dropdown.
3. Choose one of:
   - **CPU** — Standard option for most tasks (default).
   - **T4 GPU** — Great for machine learning and image processing (free tier available).
   - **TPU** — Tensor Processing Unit, for advanced deep learning.
4. Click **Save**.

## Section 5 — The Play Button (Running Cells)

The **play button** (▶) is how you execute a code cell.

**Ways to run a cell:**
- Click the ▶ **play button** on the left side of the cell.
- Press `Shift + Enter` — runs the current cell and moves to the next one.
- Press `Ctrl + Enter` — runs the current cell and stays on it.

**What the icons mean:**
- **Spinning circle** — the cell is currently running.
- ✅ **Green checkmark** — the cell ran successfully.
- ❌ **Red X** — there was an error. Read the error message below the cell.
- **[5]** — the number shows the order in which cells were run.

**Run all cells at once:**
- Go to **Runtime → Run all** (or press `Ctrl + F9`).

## Section 7 — Code Commenting

Code commenting is an important element for reproducibility and sharing of code. Think of comments like a lab notebook or a guide for your future self others to make it easier to execute the code.

### Commenting in Code Cells
In Python code, use `#` for single-line comments:
```python
# This is a comment — Python ignores this line
x = 5  # You can also add comments at the end of a line
```

For multi-line comments, use triple quotes:
```python
"""
This is a
multi-line comment.
"""
```

### Leaving Comments on Cells (Collaboration)
You can also leave review comments on any cell (like Google Docs comments):
1. Click the **comment icon** (💬) that appears when you hover over a cell.
2. Type your comment and press `Ctrl + Enter` to submit.
3. Collaborators will see your comments when viewing the notebook.
